```python
import torch
import torch.nn as nn
linear = nn.Linear(in_features = 4096, out_features=10) #transforma las dimensiones del tensor, una capa linear multiplica por una matriz. solamente toma tensores de dimension 2, tambien se llama fully connected
conv = nn.Conv2d(in_channels = 3, out_channels=9, kernel_size=3, stride=1, padding=1) #Modifica los canales pero no modifica el tama√±o de la imagen
relu = nn.ReLU(False) #rectified linear unit, en output un nuevo tendor y no el original
#las capas son funciones, 
```

```python
 x = torch.randn(4096)
 y = linear(x)
 y.size()
```

```python
x = torch.randn(1,3,7,7) # una imagen, 3 canales, dimensiones 7x7
y = conv(x)
z = relu(x)
```

```python
y.size() #solo modifico los canales de 3 a 9
```
